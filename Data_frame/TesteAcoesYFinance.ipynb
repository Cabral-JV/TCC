```python
import yfinance as yf

def verificar_cotacao_acoes(acoes, data_inicio, data_fim):
    acoes_disponiveis = []
    acoes_indisponiveis = []

    for acao in acoes:
        try:
            # Tenta obter os dados do histórico de preços para a ação no período especificado
            hist = yf.download(acao, start=data_inicio, end=data_fim)
            if hist.empty:
                acoes_indisponiveis.append(acao)
            else:
                acoes_disponiveis.append(acao)
        except Exception as e:
            acoes_indisponiveis.append(acao)

    return acoes_disponiveis, acoes_indisponiveis

# Exemplo de uso
acoes = ['AALR3.SA', 'ABCB3.SA', 'ABCB4.SA', 'ABEV3.SA', 'ABRE3.SA', 'ABYA3.SA', 'ACES3.SA', 'ACES4.SA', 'ADHM3.SA']  # Lista de ações a verificar
data_inicio = "2017-01-01"  # Data de início do período
data_fim = "2022-12-31"  # Data de fim do período

acoes_disponiveis, acoes_indisponiveis = verificar_cotacao_acoes(ativos, data_inicio, data_fim)

# Exibindo os resultados
print("\nAções disponíveis:")
for acao in acoes_disponiveis:
    print(acao)

print("\nAções indisponíveis:")
for acao in acoes_indisponiveis:
    print(acao)

# Contagem dos resultados
total_acoes_disponiveis = len(acoes_disponiveis)
total_acoes_indisponiveis = len(acoes_indisponiveis)

print("\nResultados:")
print(f"Total de ações disponíveis: {total_acoes_disponiveis}")
print(f"Total de ações indisponíveis: {total_acoes_indisponiveis}")
```
